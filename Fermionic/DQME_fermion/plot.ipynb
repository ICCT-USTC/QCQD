```python
import numpy as np
import matplotlib.pyplot as plt

from data import rho_tilde, rdo_anderson
from qiskit.quantum_info import Statevector

```

```python
rdo = np.load(file='rdo1.npz')
pop = np.empty((8,200))
for i in range(200):
    pop[0,i] = (rdo[str(i)][0,0]).real
    pop[1,i] = (rdo[str(i)][1,1]).real
    pop[2,i] = (rdo[str(i)][2,2]).real
    pop[3,i] = (rdo[str(i)][3,3]).real
    pop[4,i] = (rdo[str(i)][0,3]).real
    pop[5,i] = (rdo[str(i)][3,0]).real
    # pop[4,i] = (rdo[str(i)][0,0]).imag
    # pop[5,i] = (rdo[str(i)][1,1]).imag
    # pop[6,i] = (rdo[str(i)][2,2]).imag
    # pop[7,i] = (rdo[str(i)][3,3]).imag

plt.plot(np.arange(200) * 0.05, pop[0], label='n00real', color='red')
# plt.plot(np.arange(200) * 0.05, pop[4], label='n00imag', color='darkred', linestyle='dashed')
plt.plot(np.arange(200) * 0.05, pop[1], label='n11real', color='blue')
# plt.plot(np.arange(200) * 0.05, pop[5], label='n11imag', color='darkblue', linestyle='dashed')
plt.plot(np.arange(200) * 0.05, pop[2], label='n22real', color='green')
# plt.plot(np.arange(200) * 0.05, pop[6], label='n22imag', color='darkgreen', linestyle='dashed')
plt.plot(np.arange(200) * 0.05, pop[3], label='n33real', color='violet')
# plt.plot(np.arange(200) * 0.05, pop[7], label='n33imag', color='purple', linestyle='dashed')
plt.plot(np.arange(200) * 0.05, pop[4], label='n03real', color='black')
plt.plot(np.arange(200) * 0.05, pop[5], label='n30real', color='yellow')
plt.title("anderson model", fontsize=30, loc='center', color='purple')
plt.legend(loc='best')
plt.xlabel("Time"), plt.ylabel("Population")
# plt.savefig('nd_qc4.pdf')
plt.show()
plt.close()
```
