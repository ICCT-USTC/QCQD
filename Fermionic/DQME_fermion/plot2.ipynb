```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rdo = np.load(file='rdo.npz')
pop = np.empty((8,200))
for i in range(200):
    pop[0,i] = (rdo[str(i)][0,0]).real
    # pop[1,i] = (rdo[str(i)][0,0]).imag
    pop[1,i] = (rdo[str(i)][0,1]).real
    pop[2,i] = (rdo[str(i)][1,0]).real
    pop[3,i] = (rdo[str(i)][1,1]).real
    pop[4,i] = (rdo[str(i)][0,1]).imag
    pop[5,i] = (rdo[str(i)][1,0]).imag

plt.plot(np.arange(200) * 0.05, pop[0], label='n00real', color='red')
plt.plot(np.arange(200) * 0.05, pop[1], label='n01real', color='blue')
plt.plot(np.arange(200) * 0.05, pop[2], label='n10real', color='green')
plt.plot(np.arange(200) * 0.05, pop[3], label='n11real', color='violet')
plt.plot(np.arange(200) * 0.05, pop[4], label='n01imag', color='darkred', linestyle='dashed')
plt.plot(np.arange(200) * 0.05, pop[5], label='n10imag', color='purple', linestyle='dashed')
plt.title("anderson model", fontsize=30, loc='center', color='purple')
plt.legend(loc='best')
plt.xlabel("Time"), plt.ylabel("Population")
# plt.savefig('sys1_multimodes_qc1.pdf')
plt.show()
plt.close()
```

```python
heom_data = np.loadtxt("prop-rho-eq.dat")
pop_heom = np.empty((8,100))
for i in range(100):
    pop_heom[0,i] = heom_data[i,1]
    pop_heom[1,i] = heom_data[i,2]
    pop_heom[2,i] = heom_data[i,7]
    pop_heom[3,i] = heom_data[i,8]
plt.plot(np.arange(100) * 0.1, pop_heom[0], label='n00real', color='red')
plt.plot(np.arange(100) * 0.1, pop_heom[1], label='n00imag', color='darkred', linestyle='dashed')
plt.plot(np.arange(100) * 0.1, pop_heom[2], label='n11real', color='violet')
plt.plot(np.arange(100) * 0.1, pop_heom[3], label='n11imag', color='purple', linestyle='dashed')
plt.title("anderson model", fontsize=30, loc='center', color='purple')
plt.legend(loc='best')
plt.xlabel("Time"), plt.ylabel("Population")
# plt.savefig('sys1_mutimodes_heom1.pdf')
plt.show()
plt.close() 
```
